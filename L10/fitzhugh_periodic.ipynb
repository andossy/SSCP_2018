```python
import numpy as np
k = 2.0
A = 1.0
alpha = 0.1
L =100
eps = 0.005
gamma = 2.0

dx = 1;
dt = 0.1;

N = int(L/dx)

V = np.zeros(N);
W = np.zeros(N);
mid =  int(N/2);
V[mid-10:mid+10] = 0.3;
W[:int(0.9*mid)] = 0.1
#left =  int(N/10);
#V[:left] = 0.3;

I = np.arange(N)
Ip = I+1
Ip[N-1] = 0
Im = I-1;
Im[0] = N-1
```

```python
%matplotlib inline
import time
import matplotlib.pyplot as pl
from IPython import display
```

```python
for i in range(2000):
    
    V[I] = V[I] + dt*(k/dx**2)*(V[Ip]-2*V[I] +V[Im])
    #V[0] = V[1]
    #V[N-1] = V[N-2]
    I_ion = A*V*(1-V)*(V-alpha) -W
    
    V = V + dt*I_ion;
    W = W +dt*eps*(V-gamma*W)
    if i%20==0:
        pl.clf()
        pl.axis([0, L, 0, 1.1])
        pl.plot(V)
        pl.plot(W)
        pl.title('i=%d' % i)
        display.clear_output(wait=True)
        display.display(pl.gcf())
```
